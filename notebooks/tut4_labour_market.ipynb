```julia
function labour_demand(A = 20, K = 20, w = 1:50, α = 0.6)
    Nd = K .* ((1 .- α) .* A ./ w) .^ (1 .- α)
end;
```

Plot this labour demand function

```julia
Nd = labour_demand();
```

```julia
plot(Nd) # gives us a labour demand function
```
